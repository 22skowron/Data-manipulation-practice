```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')

tips.info()
tips.nunique()
```

```python
tips.sample(3)
```

```python
tips.describe()
```

```python
tips['time'].value_counts()
```

```python
sns.set_theme()

sns.relplot(
    data=tips,
    x='total_bill', y='tip', col='sex',
    hue='smoker', style='smoker'
)
```

```python
rng = np.random.default_rng()
```

```python
v1 = rng.normal(0, 1, size=3000)
v2 = rng.normal(0, 5, size=3000)
v3 = rng.normal(10, 3, size=3000)
```

```python
fig, ax1 = plt.subplots(layout='constrained', figsize=(12, 4))

sns.histplot(x=v1, kde=True, bins=10, ax=ax1, label='N(0, 1)')
sns.histplot(x=v2, kde=True, bins=20, ax=ax1, label='N(0, 5)')
sns.histplot(x=v3, kde=True, bins=20, ax=ax1, label='N(10, 3)')

# ax1.set_xlim(-50, 50)
ax1.legend()
```
